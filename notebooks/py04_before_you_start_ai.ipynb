## Quick HuggingFace tour

You'll see this pattern in almost every AI phase. Run it now so it's not new when you get there.

````python
from transformers import AutoTokenizer

# Load a tokenizer for GPT-2
tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "Transformers changed everything in AI."
tokens = tokenizer(text)

print("Text:     ", text)
print("Token IDs:", tokens['input_ids'])
print("Decoded:  ", tokenizer.decode(tokens['input_ids']))
print(f"\n{len(tokens['input_ids'])} tokens for {len(text.split())} words")
````

## Glossary

Terms you'll encounter constantly in the AI phases — one sentence each.

| Term | What it means |
|---|---|
| **Token** | A chunk of text — a word, part of a word, or punctuation. "Hello" might be 1 token; "tokenization" might be 3. |
| **Embedding** | A vector (list of numbers) that represents a token or sentence. Similar meanings → similar vectors. |
| **Attention** | A mechanism that lets a model focus on the most relevant parts of the input when processing each token. |
| **Transformer** | The architecture behind GPT, BERT, and every modern LLM. Built entirely from attention + feed-forward layers. |
| **Pre-training** | Training a model on a huge dataset to learn general language understanding. Very expensive, done once. |
| **Fine-tuning** | Taking a pre-trained model and training it further on a smaller, task-specific dataset. |
| **LoRA** | Low-Rank Adaptation — a technique to fine-tune a model by training only a tiny fraction of its weights. |
| **RAG** | Retrieval-Augmented Generation — giving a model access to external documents at inference time. |
| **Inference** | Running a trained model to make predictions. Training is learning; inference is using what was learned. |
| **Loss** | A number measuring how wrong the model's predictions are. Training tries to minimise it. |
| **Gradient** | The direction and magnitude of change needed to reduce the loss. Computed by autograd. |
| **Batch** | A group of training examples processed together. Typical batch sizes: 8, 16, 32, 64. |
| **Epoch** | One full pass through the training dataset. |
| **Checkpoint** | A saved snapshot of a model's weights at a point in training. |

## Key papers (optional reading)

These are the papers that built the field. You don't need to read them to follow the course — but knowing they exist is useful context.

| Year | Paper | Why it matters |
|---|---|---|
| 2017 | [Attention Is All You Need](https://arxiv.org/abs/1706.03762) | Introduced the Transformer architecture — the foundation of every modern LLM |
| 2018 | [BERT](https://arxiv.org/abs/1810.04805) | Showed that pre-training on text + fine-tuning beats task-specific models |
| 2019 | [GPT-2](https://openai.com/research/better-language-models) | Demonstrated that large language models can generate coherent long-form text |
| 2020 | [GPT-3](https://arxiv.org/abs/2005.14165) | Showed that scale alone produces few-shot learning — no fine-tuning needed |
| 2021 | [LoRA](https://arxiv.org/abs/2106.09685) | Parameter-efficient fine-tuning — train 0.1% of weights, get 99% of the performance |
| 2020 | [RAG](https://arxiv.org/abs/2005.11401) | Combined retrieval with generation — models that can look things up |
