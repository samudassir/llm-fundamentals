# PY 02 — Matplotlib for AI

AI code produces a lot of numbers. Matplotlib turns those numbers into pictures you can actually understand.

By the end you'll be able to:
- Plot training loss curves
- Visualize token distributions
- Plot attention heatmaps
- Visualize embeddings in 2D

These are the exact plots you'll see throughout the AI phases.

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
print("Ready")
```

---
## 1. The basics — line plots

The most common plot in AI: the **loss curve**. It shows whether your model is learning.

```python
# Simulate a training loss curve
epochs = np.arange(1, 51)
loss   = 2.5 * np.exp(-0.08 * epochs) + 0.1 + np.random.randn(50) * 0.05

plt.figure(figsize=(8, 4))
plt.plot(epochs, loss, color='steelblue', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Train vs validation loss — the most important plot in ML
# Gap between them = overfitting
train_loss = 2.5 * np.exp(-0.09 * epochs) + 0.08 + np.random.randn(50) * 0.03
val_loss   = 2.5 * np.exp(-0.07 * epochs) + 0.25 + np.random.randn(50) * 0.05

plt.figure(figsize=(8, 4))
plt.plot(epochs, train_loss, label='Train',      color='steelblue', linewidth=2)
plt.plot(epochs, val_loss,   label='Validation', color='tomato',    linewidth=2, linestyle='--')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train vs Validation Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print("If val loss rises while train loss falls → overfitting")
```

---
## 2. Bar charts — token frequencies

Used to visualize which tokens appear most often in a dataset.

```python
tokens = ['the', 'a', 'is', 'in', 'it', 'of', 'to', 'and', 'was', 'he']
counts = [1200, 980, 870, 760, 650, 620, 590, 540, 480, 430]

plt.figure(figsize=(9, 4))
plt.bar(tokens, counts, color='steelblue', edgecolor='white')
plt.xlabel('Token')
plt.ylabel('Count')
plt.title('Most Frequent Tokens')
plt.tight_layout()
plt.show()
```

---
## 3. Histograms — distributions

Used to understand the distribution of weights, activations, or any numerical values in a model.

```python
# Model weights are usually normally distributed
weights = np.random.randn(10000) * 0.02  # typical weight initialisation

plt.figure(figsize=(8, 4))
plt.hist(weights, bins=60, color='steelblue', edgecolor='white', alpha=0.8)
plt.xlabel('Weight value')
plt.ylabel('Count')
plt.title('Distribution of Model Weights')
plt.axvline(0, color='tomato', linestyle='--', label='zero')
plt.legend()
plt.tight_layout()
plt.show()
```

---
## 4. Heatmaps — attention weights

The most important visualization in Transformer models. Attention heatmaps show which tokens are "attending" to which other tokens.

```python
# Simulate attention weights for a short sentence
tokens = ['The', 'cat', 'sat', 'on', 'the', 'mat']
n = len(tokens)

# Fake attention weights — rows attend to columns
np.random.seed(42)
attn = np.random.rand(n, n)
# Make it look realistic — each row sums to 1 (softmax output)
attn = attn / attn.sum(axis=1, keepdims=True)

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(attn, cmap='Blues')
ax.set_xticks(range(n)); ax.set_xticklabels(tokens, rotation=45, ha='right')
ax.set_yticks(range(n)); ax.set_yticklabels(tokens)
ax.set_title('Attention Weights')
ax.set_xlabel('Key (attended to)')
ax.set_ylabel('Query (attending)')
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
print("Darker = stronger attention. You'll see real versions of this in AI Phase 01.")
```

## 5. Scatter plots — embedding visualization

Word embeddings live in high-dimensional space (512, 768 dimensions). We compress them to 2D to visualize which words are close to each other.

```python
# Simulate 2D word embeddings (in reality you'd use PCA or t-SNE to reduce dims)
np.random.seed(7)

# Words clustered by meaning
royalty  = {'words': ['king', 'queen', 'prince', 'princess'],
             'xy': np.random.randn(4, 2) * 0.3 + [2, 2]}
animals  = {'words': ['cat', 'dog', 'bird', 'fish'],
             'xy': np.random.randn(4, 2) * 0.3 + [-2, 2]}
vehicles = {'words': ['car', 'bus', 'train', 'plane'],
             'xy': np.random.randn(4, 2) * 0.3 + [0, -2]}

fig, ax = plt.subplots(figsize=(7, 6))
for group, color in [(royalty, 'steelblue'), (animals, 'tomato'), (vehicles, 'seagreen')]:
    ax.scatter(group['xy'][:, 0], group['xy'][:, 1], color=color, s=100, zorder=3)
    for word, (x, y) in zip(group['words'], group['xy']):
        ax.annotate(word, (x, y), textcoords='offset points', xytext=(6, 4), fontsize=11)

ax.set_title('Word Embeddings in 2D')
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
print("Words with similar meanings cluster together.")
```

---
## 6. Subplots — multiple plots together

You'll often want to show several plots side by side.

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Plot 1 — loss curve
axes[0].plot(epochs, train_loss, color='steelblue')
axes[0].set_title('Training Loss')
axes[0].set_xlabel('Epoch')
axes[0].grid(True, alpha=0.3)

# Plot 2 — weight distribution
axes[1].hist(np.random.randn(1000), bins=30, color='tomato', alpha=0.8)
axes[1].set_title('Weight Distribution')
axes[1].set_xlabel('Value')

# Plot 3 — attention (mini)
mini_attn = np.random.rand(4, 4)
mini_attn = mini_attn / mini_attn.sum(axis=1, keepdims=True)
axes[2].imshow(mini_attn, cmap='Blues')
axes[2].set_title('Attention Weights')

plt.suptitle('Common AI Visualizations', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```
